# Binary Tree Inorder Traversal

Given the root of a binary tree, return the inorder traversal of its nodes' values.

Example 1:
Input: root = [1,null,2,3]
Output: [1,3,2]

Constraints:

The number of nodes in the tree is in the range [0, 100].
    -100 <= Node.val <= 100


## Solution

```kotlin
data class TreeNode(
    val `val`: Int,
    var left: TreeNode? = null,
    var right: TreeNode? = null
)

// Convert an array (with nulls) into a binary tree
fun buildTree(arr: Array<Int?>): TreeNode? {
    if (arr.isEmpty() || arr[0] == null) return null

    val root = TreeNode(arr[0]!!)
    val queue: ArrayDeque<TreeNode> = ArrayDeque()
    queue.add(root)

    var i = 1
    while (i < arr.size) {
        val current = queue.removeFirst()

        // Left child
        if (i < arr.size && arr[i] != null) {
            current.left = TreeNode(arr[i]!!)
            queue.add(current.left!!)
        }
        i++

        // Right child
        if (i < arr.size && arr[i] != null) {
            current.right = TreeNode(arr[i]!!)
            queue.add(current.right!!)
        }
        i++
    }

    return root
}
```

```kotlin
class Solution {
    fun inorderTraversal(root: TreeNode?): List<Int> {
        val result = mutableListOf<Int>()

        fun dfs(root: TreeNode?) {
            if (root == null) return

            dfs(root.left)
            result.add(root.`val`)
            dfs(root.right)
        }

        dfs(root)
        return result
    }
}
```

Example 1:

```kotlin
fun main() {
    val rootNode = buildTree(arrayOf(1,2,3,4,5,null,8,null,null,6,7,9))
    println(
        Solution().inorderTraversal(rootNode)
    )
}
main()
```
